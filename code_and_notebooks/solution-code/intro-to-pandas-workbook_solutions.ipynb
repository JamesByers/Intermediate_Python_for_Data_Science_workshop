<h3>Blooper alert: The best way to create a DataFrame is to use a dictionary. But, remember to keep the list of values the SAME LENGTH.</h3>

```python
mydata = pd.DataFrame({'Letters':['A','B','C'], 'Integers':[1,2,3], 'Floats':[2.2, 3.3, 4.4]}, index=mydata['Letters'])
mydata
```

<h3>Blooper alert: Don't forget to add `inplace=True`, otherwise it won't work.</h3>

```python
mydata.rename(columns={'Integers':'Ints'},inplace=True)
mydata
```

```python
mydata.dtypes, mydata.columns
```

```python
mydata.columns=['A','B','C']
mydata
```

```python
mydata.ix[1,'B'] = 100
mydata
```

```python
mydata.ix[:,'A'] = 'foo'
mydata
```

```python
mydata.ix[0,['B','C']]  = [1000,1]
mydata
```
